```
import networkx as nx
import matplotlib.pyplot as plt


#Create graph

GoT = nx.Graph() 

#Add characters as nodes

GoT.add_nodes_from(['Daenarys','Howland','AerysII','Jaime','Viserys','Lyanna',
                    'Elia','Robert','JonArryn','RhaenysNAegon','Ned','Gregor',
                    'Rhaella','Rickard','Arthur','JonSnow','Benjen','Rhaegar',
                    'Brandon','Tywin'])

#### Create abduction graph

Abduct = nx.DiGraph()
Abduct.add_edges_from([('Rhaegar','Lyanna')], label='abducted')
Abduct_labels = nx.get_edge_attributes(Abduct, 'label')


#### Create Serve graph

Serve = nx.DiGraph()
Serve.add_edges_from([('Arthur','AerysII'),('Jaime','AerysII'),('Gregor','Tywin')], label='Serves')
Serve_labels = nx.get_edge_attributes(Serve, 'label')


#### Create Guardian graph

Guard = nx.DiGraph()
Guard.add_edges_from([('Ned','JonSnow'),('JonArryn','Robert'),('JonArryn','Ned')], label='Guards')
Guard_labels = nx.get_edge_attributes(Guard, 'label')


#### Create Kill graph

Kill = nx.DiGraph()
Kill.add_edges_from([('Ned','Arthur'),('Gregor','RhaenysNAegon'),('Gregor','Elia'),
                     ('Howland','Arthur'),('AerysII','Brandon'),
                     ('AerysII','Rickard'),('Jaime','AerysII'),('Robert','Rhaegar')], label='Kills')
Kill_labels = nx.get_edge_attributes(Kill, 'label')


#### Create Marriage Graph

Married = nx.Graph()
Married.add_edges_from([('Lyanna','Robert'),('Elia','Rhaegar'),('Rhaella','AerysII')],label="Married")
Marriage_labels = nx.get_edge_attributes(Married, 'label')

#### Create Allies Graph

Allies = nx.Graph()
Allies.add_edges_from([('Tywin','Robert'),('Robert','Ned'),('Robert','JonArryn'),
                       ('Ned','Howland'),('Robert','JonArryn')],label="Allies")
Allies_labels = nx.get_edge_attributes(Allies, 'label')


#### Create Parents graph

Parents = nx.DiGraph()
Parents.add_edges_from([('Elia','RhaenysNAegon'),('Rickard','Ned'),('Rickard','Lyanna'),
                     ('Rickard','Benjen'),('Rickard','Brandon'),
                     ('Rhaegar','JonSnow'),('Rhaegar','RhaenysNAegon'),('Rhaegar','Rhaella'),
                     ('Tywin','Jaime'),('Rhaella','Rhaegar'),('Rhaella','Daenarys'),
                     ('AerysII','Daenarys'),('AerysII','Rhaegar'),('AerysII','Viserys'),
                     ('Lyanna','JonSnow')], label='Parent')
Parents_labels = nx.get_edge_attributes(Parents , 'label')


#### Create Siblings Graph

Siblings = nx.Graph()
Siblings.add_edges_from([('Lyanna','Ned'),('Lyanna','Benjen'),('Lyanna','Brandon'),
                       ('Ned','Brandon'),('Ned','Benjen'),('Rhaella','AerysII'),
                       ('Benjen','Brandon'),('Rhaegar','Viserys'),('Rhaegar','Daenarys'),
                       ('Daenarys','Viserys')],label="Siblings")
Siblings_labels = nx.get_edge_attributes(Siblings, 'label')


plt.figure(1, figsize=(12,12))
pos=nx.circular_layout(GoT)

nx.draw(GoT,  pos, node_color='grey', node_size=5, with_labels=True)

nx.draw_networkx_edges(Abduct, pos, edge_color='pink')
nx.draw_networkx_edge_labels(Abduct, pos, edge_labels=Abduct_labels, label_pos=0.3)

nx.draw_networkx_edges(Serve, pos, edge_color='blue')
nx.draw_networkx_edge_labels(Serve, pos, edge_labels=Serve_labels, label_pos=0.3)

nx.draw_networkx_edges(Guard, pos, edge_color='lightblue')
nx.draw_networkx_edge_labels(Guard, pos, edge_labels=Guard_labels, label_pos=0.3)

nx.draw_networkx_edges(Kill, pos, edge_color='green')
nx.draw_networkx_edge_labels(Kill, pos, edge_labels=Kill_labels, label_pos=0.3)

nx.draw_networkx_edges(Married, pos, edge_color='deeppink')
nx.draw_networkx_edge_labels(Married, pos, edge_labels=Marriage_labels, label_pos=0.6)

nx.draw_networkx_edges(Allies, pos, edge_color='yellow')
nx.draw_networkx_edge_labels(Allies, pos, edge_labels=Allies_labels, label_pos=0.6)

nx.draw_networkx_edges(Parents, pos, edge_color='black')
nx.draw_networkx_edge_labels(Parents, pos, edge_labels=Parents_labels, label_pos=0.6)


nx.draw_networkx_edges(Siblings, pos, edge_color='purple')
nx.draw_networkx_edge_labels(Siblings, pos, edge_labels=Siblings_labels, label_pos=0.6)


print('most action in general')  
for n in GoT.nodes():
  print(n,':',Abduct.degree(n),Serve.degree(n),Guard.degree(n),Kill.degree(n),Married.degree(n),Allies.degree(n),Parents.degree(n),Siblings.degree(n) )

  
print('\nmost action initiated')  
for n in GoT.nodes():
  print(n,':',Abduct.out_degree(n),Serve.out_degree(n),Guard.out_degree(n),Kill.out_degree(n),Parents.out_degree(n) )

 
print('\nRhaegar has the most eventful story with 10 degrees\nAerysII initiated the most action in the story with 5 degrees')

```
